```python
# 2. Buatlah sebuah dataframe yang menunjukkan data usia maksimal, minimal & rata-ratanya, berdasarkan profesi & gender!
```

```python
dp4 = dp.groupby([dp['occupation'],dp['gender']]).describe()

dp4 = dp4['age'][['max','min','mean']]

dp4.rename(columns={'max':'max_usia','min':'min_usia','mean':'rerata_usia'}, inplace=True)

dp4
```

```python
# 3. Buatlah sebuah dataframe yang menunjukkan persentase pria & wanita tiap profesi!
```

```python
dp5 = pd.crosstab(dp.occupation, dp.gender).apply(lambda i: i/i.sum(), axis= 1) * 100

dp5['%total'] = dp5['F']+dp5['M']

dp5.rename(columns={'F':'%female','M':'%male'},inplace=True)

dp5 = dp5[['%male','%female','%total']]

dp5
```
